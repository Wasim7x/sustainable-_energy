### Description
- this dataset showcasing sustainable energy indicators and other useful factors across all countries from 2000 to 2020. 

- It includes data such as electricity access, renewable energy, carbon emissions, energy intensity, Financial flows, and economic growth. 

- It help to compare nations, track progress towards **Sustainable Development** and gain profound insights into **global energy consumption** patterns over time.







### Key Features:

- **Entity**:   The name of the country or region for which the data is reported.

- **Year**:   The year for which the data is reported, ranging from 2000 to 2020.

- **Access to electricity (% of population)**:   The percentage of population with access to electricity.

- **Access to clean fuels for cooking (% of population)**:   The percentage of the population with primary reliance on clean fuels.

- **Renewable-electricity-generating-capacity-per-capita**:   Installed Renewable energy capacity per person.

- **Financial flows to developing countries (US $)**:   Aid and assistance from developed countries for clean energy projects.


- **Renewable energy share in total final energy consumption (%)**:   Percentage of renewable energy in final energy consumption.

- **Electricity from fossil fuels (TWh)**:   Electricity generated from fossil fuels (coal, oil, gas) in terawatt-hours.

- **Electricity from nuclear (TWh)**:   Electricity generated from nuclear power in terawatt-hours.

- **Electricity from renewables (TWh)**:   Electricity generated from renewable sources (hydro, solar, wind, etc.) in terawatt-hours.

- **Low-carbon electricity (% electricity)**:   Percentage of electricity from low-carbon sources (nuclear and renewables).

- **Primary energy consumption per capita (kWh/person)**:   Energy consumption per person in kilowatt-hours.

- **Energy intensity level of primary energy (MJ/$2011 PPP GDP)**: Energy use per unit of GDP at purchasing power parity.

- **Value_co2_emissions (metric tons per capita)**:   Carbon dioxide emissions per person in metric tons.

- **Renewables (% equivalent primary energy)**:   Equivalent primary energy that is derived from renewable sources.

- **GDP growth (annual %)**:   Annual GDP growth rate based on constant local currency.

- **GDP per capita**:   Gross domestic product per person.

- **Density (P/Km2)**:   Population density in persons per square kilometer.

- **Land Area (Km2)**:   Total land area in square kilometers.



### Use cases

- **Energy Consumption Prediction**: Predict future energy usage

- **Carbon Emission Forecasting**: Forecast CO2 emissions, support climate strategies.

- **Energy Equity Analysis**: Analyze access, density, and growth for equitable distribution.

- **Energy Efficiency Optimization**: Identify intensive areas for environmental impact reduction.

- **Renewable Energy Potential Assessment**: Identify regions for green investments based on capacity.

- **Renewable Energy Investment Strategies**: Guide investors towards sustainable opportunities.

### Feature Engineering

```python
# Stander scaler use for scaling the data
from sklearn.preprocessing import StandardScaler
 
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)
df = pd.DataFrame(scaled_data, columns=df.columns)
```

## Modeling

#### Training and Test dataset division

```python
from sklearn.preprocessing import LabelEncoder # for level encoding of categorical value
df['Entity'] = LabelEncoder().fit_transform(df.Entity.values) #converting str data(country/Entity name) to numeric value using lable encoding
```

```python
from sklearn.model_selection import train_test_split # use to splite data into training and test set 
from sklearn.metrics import root_mean_squared_error, mean_absolute_error,r2_score # use to test modale accuracy 
# Assuming 'co2 emission' is target variable
X = df.drop('Value_co2_emissions_kt_by_country', axis=1)
y = df['Value_co2_emissions_kt_by_country']
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25)   
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression # linear Regg class (prediction)
linearReg_model = LinearRegression() # linear Regg modle object

linearReg_model.fit(X_train,y_train) # model trainig

predictions_linear = linearReg_model.predict(X_test) # model prediction

# Evaluate the model
mse_linear = root_mean_squared_error(y_test, predictions_linear)  # Mean Squared Error (error term)
mae_linear = mean_absolute_error(y_test, predictions_linear) # Mean absolute Error
r2_linear = r2_score(y_test, predictions_linear)    # (modal accuracy)        

# Print evaluation metrics
print(f"Model: {type(linearReg_model).__name__}, mse: {mse_linear}")
print(f"Model: {type(linearReg_model).__name__}, mae: {mae_linear}")
print(f"Model: {type(linearReg_model).__name__}, r2: {r2_linear}")
```

#### Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
# Make predictions
model_prediction = model.predict(X_test)

# Evaluate the model
mse_linear = root_mean_squared_error(y_test, model_prediction) # Mean Squared Error
mae_linear = mean_absolute_error(y_test, model_prediction)
r2_linear = r2_score(y_test, model_prediction)

# Print evaluation metrics
print(f"Model: {type(model).__name__}, mse: {mse_linear}")
print(f"Model: {type(model).__name__}, mae: {mae_linear}")
print(f"Model: {type(model).__name__}, r2: {r2_linear}")
```

#### KNearest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
model.fit(X_train,y_train)
# Make predictions
model_prediction = model.predict(X_test)

# Evaluate the model
mse_linear = root_mean_squared_error(y_test, model_prediction) # Mean Squared Error
mae_linear = mean_absolute_error(y_test, model_prediction)
r2_linear = r2_score(y_test, model_prediction)

# Print evaluation metrics
print(f"Model: {type(model).__name__}, mse: {mse_linear}")
print(f"Model: {type(model).__name__}, mae: {mae_linear}")
print(f"Model: {type(model).__name__}, r2: {r2_linear}")
```

#### Model Comparison

#### compair model on the basis of mse_score

```python
from sklearn.model_selection import cross_val_score
# List of algorithms to compare
algorithms = [
    LinearRegression(),
    DecisionTreeRegressor(),
    KNeighborsRegressor()
]

largest_mse = float('inf') # it assingn largest possible no. of flote datatype
best_model = None

# Loop through each algorithm
for model in algorithms:
    # train the model
    model.fit(X_train, y_train)

    # Make predictions
    predictions = model.predict(X_test)

    # model evaluation using cross-validation: mean squared error
    mse_scores = -cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error') 
    mean_mse = np.mean(mse_scores)  # geting mean of mse_scors given by cross validation score
    
    # Update the best model if the current model has lower mean squared error
    if mean_mse < largest_mse:
        largest_mse = mean_mse
        best_model = model

# Print the best model and its mean squared error
print("\nBest Model:")
print(best_model)
print("Best Cross-Validation MSE:", largest_mse)
```

#### compair model on the basis of mae_score

```python
# List of algorithms to compare
algorithms = [
    LinearRegression(),
    DecisionTreeRegressor(),
    KNeighborsRegressor()
]

largest_mae = float('inf') # it assingn largest possible no. of flote datatype
best_model = None

# Loop through each algorithm
for model in algorithms:
    # train the model
    model.fit(X_train, y_train)

    # Make predictions
    predictions = model.predict(X_test)

    # Evaluate the model using cross-validation with mean squared error
    mae_scores = -cross_val_score(model, X, y, cv=5, scoring='neg_mean_absolute_error')
    mean_mae = np.mean(mae_scores)  # geting mean of mse_scors given by cross validation score
    
    # Update the best model if the current model has lower mean squared error
    if mean_mae < largest_mae:
        largest_mae = mean_mae
        best_model = model

# Print the best model and its mean squared error
print("\nBest Model:")
print(best_model)
print("Best Cross-Validation MAE:", largest_mae)
```

#### compair model on basis of r2_score

```python
# List of algorithms to compare
algorithms = [
    LinearRegression(),
    DecisionTreeRegressor(),
    KNeighborsRegressor()
]

largest_r2 = float('-inf')  # Start with the smallest possible value
best_model = None

# Loop through each algorithm
for model in algorithms:
    # Train the model
    model.fit(X_train, y_train)

    # Evaluate the model using cross-validation with R^2 score
    r2_scores = cross_val_score(model, X, y, cv=5, scoring='r2')
    mean_r2 = np.mean(r2_scores)  # Getting the mean R^2 score from cross-validation

    # Update the best model if the current model has a higher R^2 score
    if mean_r2 > largest_r2:
        largest_r2 = mean_r2
        best_model = model

# Print the best model and its R^2 score
print("\nBest Model:")
print(best_model)
print("Best Cross-Validation R^2:", largest_r2)
```
