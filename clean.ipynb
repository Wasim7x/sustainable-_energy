#### Data cleanig

```python
# classify the col of num and cat datatype
numeric_data = df.select_dtypes(include='number') #numaric data values
numeric_col = df.select_dtypes(include='number').columns # numaric data col
print(numeric_col)
categorical_data = df.select_dtypes(exclude='number') #categorical data
categorical_col = df.select_dtypes(exclude='number').columns #categorical data col 
print(categorical_col)
```

```python
# Drop columns with a high number of missing values
df.drop(columns=['Financial flows to developing countries (US $)', 'Renewables (% equivalent primary energy)',
                 'Renewable-electricity-generating-capacity-per-capita'], inplace=True) # inplace = true - permanent change in df
                                                                                        
```

```python
# Fill missing values with mean
columns_to_fill_mean = ['Access to clean fuels for cooking', 'Renewable energy share in the total final energy consumption (%)',
                        'Electricity from nuclear (TWh)', 'Energy intensity level of primary energy (MJ/$2017 PPP GDP)',
                        'Value_co2_emissions_kt_by_country', 'gdp_growth', 'gdp_per_capita',]
df[columns_to_fill_mean] = df[columns_to_fill_mean].apply(lambda x: x.fillna(x.mean()))
```

```python
df.isna().sum() # recheck is null persent

# missing value for feature columns: Access to electricity (% of population),Electricity from fossil fuels (TWh) etc are 
# not handle by mean value as it may  mislead the data 
```

```python
# drop remaining rows with null value-these na value not replace by mean value
df = df.dropna()    
```

```python
# Check for duplicate rows
num_duplicates = df.duplicated().sum()
print(num_duplicates)
```

```python
df.isna().sum() # final check for null value
```

```python
# Rename the 'Density' column
df.rename(columns={'Density\\n(P/Km2)': 'Population_Density'}, inplace=True)
# remove    "," from density values having more than three digits  
df.loc[:, 'Population_Density'] = df['Population_Density'].astype(str).str.replace(',', '').astype(int) # 1st convert int value to str for replacing "," with blank and then convert back to int
print(df.columns)
```

## Modeling

#### Training and Test dataset division

```python
from sklearn.preprocessing import LabelEncoder # for level encoding of categorical value
df['Entity'] = LabelEncoder().fit_transform(df.Entity.values) #converting str data(country/Entity name) to numeric value using lable encoding
```

```python
from sklearn.model_selection import train_test_split # use to splite data into training and test set 
from sklearn.metrics import root_mean_squared_error, mean_absolute_error,r2_score # use to test modale accuracy 
# Assuming 'gdp_growth' is target variable
X = df.drop('gdp_growth', axis=1)
y = df['gdp_growth']
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25)   
```

#### Linear Regression

```python
from sklearn.linear_model import LinearRegression # linear Regg class
linearReg_model = LinearRegression() # linear Regg modle object

linearReg_model.fit(X_train,y_train) # model trainig

predictions_linear = linearReg_model.predict(X_test) # model prediction

# Evaluate the model
mse_linear = root_mean_squared_error(y_test, predictions_linear)  # Mean Squared Error
mae_linear = mean_absolute_error(y_test, predictions_linear) # Mean absolute Error
r2_linear = r2_score(y_test, predictions_linear)            

# Print evaluation metrics
print(f"Model: {type(linearReg_model).__name__}, mse: {mse_linear}")
print(f"Model: {type(linearReg_model).__name__}, mae: {mae_linear}")
print(f"Model: {type(linearReg_model).__name__}, r2: {r2_linear}")
```

#### Decision tree

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train,y_train)
# Make predictions
model_prediction = model.predict(X_test)

# Evaluate the model
mse_linear = root_mean_squared_error(y_test, model_prediction) # Mean Squared Error
mae_linear = mean_absolute_error(y_test, model_prediction)
r2_linear = r2_score(y_test, model_prediction)

# Print evaluation metrics
print(f"Model: {type(model).__name__}, mse: {mse_linear}")
print(f"Model: {type(model).__name__}, mae: {mae_linear}")
print(f"Model: {type(model).__name__}, r2: {r2_linear}")
```

#### KNearest Neighbors

```python
from sklearn.neighbors import KNeighborsRegressor
model = KNeighborsRegressor()
model.fit(X_train,y_train)
# Make predictions
model_prediction = model.predict(X_test)

# Evaluate the model
mse_linear = root_mean_squared_error(y_test, model_prediction) # Mean Squared Error
mae_linear = mean_absolute_error(y_test, model_prediction)
r2_linear = r2_score(y_test, model_prediction)

# Print evaluation metrics
print(f"Model: {type(model).__name__}, mse: {mse_linear}")
print(f"Model: {type(model).__name__}, mae: {mae_linear}")
print(f"Model: {type(model).__name__}, r2: {r2_linear}")
```

#### Model Comparison

```python
from sklearn.model_selection import cross_val_score
# List of algorithms to compare
algorithms = [
    LinearRegression(),
    DecisionTreeRegressor(),
    KNeighborsRegressor()
]

largest_mse = float('inf') # it assingn largest possible no. of flote datatype
best_model = None

# Loop through each algorithm
for model in algorithms:
    # train the model
    model.fit(X_train, y_train)

    # Make predictions
    predictions = model.predict(X_test)

    # model evaluation using cross-validation: mean squared error
    mse_scores = -cross_val_score(model, X, y, cv=5, scoring='neg_mean_squared_error') 
    mean_mse = np.mean(mse_scores)  # geting mean of mse_scors given by cross validation score
    
    # Print the cross-validation mean squared error
    print(f"{model.__class__.__name__} - Cross-Validation MSE: {mean_mse}")

    # Update the best model if the current model has lower mean squared error
    if mean_mse < largest_mse:
        largest_mse = mean_mse
        best_model = model

# Print the best model and its mean squared error
print("\nBest Model:")
print(best_model)
print("Best Cross-Validation MSE:", largest_mse)
```
